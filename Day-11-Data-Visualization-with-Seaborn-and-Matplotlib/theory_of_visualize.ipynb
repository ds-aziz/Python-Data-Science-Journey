```python
# Scatter plot



# x = food        

# y = weight

import matplotlib.pyplot as plt
import seaborn as sns

# Data tayar karna
food = ["biryani", "karhai", "kofta"]
weight = [50, 55, 45]  # 'kg' hata kar sirf numbers rakhe hain

# Plot banana
sns.scatterplot(x=food, y=weight, color="blue", s=100)

# Labels lagana
plt.xlabel("Favorite Food")
plt.ylabel("Weight (kg)")
plt.title("Food vs Weight Scatter Plot")

# Plot dikhana
plt.show()
```

```python
# Bar Plot

# x = food
# y = weight

import matplotlib.pyplot as plt
import seaborn as sns

# Data tayar karna
food = ["biryani", "karhai", "kofta"]
weight = [50, 55, 45]  # 'kg' hata kar sirf numbers rakhe hain

# Plot banana
sns.barplot(x=food, y=weight)

# Labels lagana
plt.xlabel("Favorite Food")
plt.ylabel("Weight (kg)")
plt.title("Food vs Weight Scatter Plot")

# Plot dikhana
plt.show()
```
