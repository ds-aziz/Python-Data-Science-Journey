```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pehle data load karna zaroori hai
kashti = sns.load_dataset("titanic")

# Ab aapka code bilkul sahi chalega
sns.stripplot(x="age", y="fare", hue="class", data=kashti)

# Plot ko screen par dikhane ke liye
plt.show()
```

```python
#heat map
# Sirf numeric data ka correlation nikalen
corr = kashti.corr(numeric_only=True)

plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm') # annot=True se numbers bhi show honge
plt.title("Heat map of kashti ka data")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data load kiya (agar pehle se loaded hai toh dubara likhne ki zaroorat nahi)
kashti = sns.load_dataset("titanic")

# 2. Sirf plot ka function likhein (variable me save MAT karein)
sns.violinplot(data=kashti, x="sex", y="fare", hue="class")

# 3. Plot dekhne ke liye
plt.show()
```

```python
# Yeh line aap ko sex aur class ke mutabiq sab se zyada (max) kiraya dikhaye gi
print(kashti.groupby(['sex', 'class'])['fare'].max())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dono ko dekhne ke liye boxplot check karein, isme outliers (dots) alag nazar aate hain
sns.boxplot(data=kashti, x='sex', y='fare', hue='class')
plt.show()
```
