# Assignment:
reason of these two things df.index (without bracket),\ Second one is df.info() with bracket.

Yeh bahut achha aur logical sawal hai. Programming mein iska seedha sa jawab hai: **"Action"** vs **"Property"**.

Python (aur Pandas) mein iska farq samajhne ke liye in dono ko aise dekho:

### 1. `df.info()` — Yeh ek "Action" (Method/Function) hai

Jab aap kisi cheez ke saath `()` lagate hain, toh aap Python se keh rahe hote hain: **"Bhai, ye kaam kar ke dikhao"** (Execute this action).

* **`df.info()`** ek function/method hai. Jab aap ise call karte hain, toh Python peeche background mein ek lambi calculation karta hai:
* Woh poori memory check karta hai.
* Column ke types dekhta hai.
* Phir ek report generate karke screen par print karta hai.


* **Reason:** Kyunki isko bahut saara kaam (computation) karna padta hai, isliye `()` ka matlab hai "Abhi process start karo".

### 2. `df.index` — Yeh ek "Property" (Attribute) hai

Jab aap bina `()` ke likhte hain, toh aap Python se keh rahe hote hain: **"Bhai, mujhe is cheez ki details la kar do"** (Give me this information).

* **`df.index`** ek property ya attribute hai. Yeh koi nayi calculation nahi kar raha, balki yeh DataFrame ke andar pehle se store ki hui ek value/list hai.
* **Reason:** Yeh ek "Static" cheez hai. DataFrame ko pata hai ki uski index kya hai, bas aapne uska data mang liya. Isliye `()` ki zaroorat nahi hoti, kyunki aap koi action nahi karwa rahe, sirf "info" check kar rahe hain.

---

### Ek Simple Example se samjho:

Socho tumhare paas ek **"Mobile Phone"** hai:

1. **`phone.call()`** (Action): Iske saath `()` lagna zaroori hai. Kyunki ye ek process hai—network dhoondhna, call milana, baat karwana.
2. **`phone.battery`** (Property): Iske saath `()` ki zaroorat nahi hai. Kyunki phone mein battery ka level pehle se "record" hai. Aap bas uska current status dekh rahe hain.

---

### Yaad rakhne ka asaan tarika:

* **`()` (Method/Function):** Jab aapko koi **kaam** karwana ho, koi analysis run karni ho, ya data change karna ho. (Example: `df.describe()`, `df.head()`, `df.dropna()`)
* **No `()` (Attribute/Property):** Jab aapko sirf **data ki value** ya koi **feature** check karna ho jo pehle se wahan mojood hai. (Example: `df.shape`, `df.columns`, `df.dtypes`)

**Ek chhoti si warning:** Agar aap `df.info` (bina brackets ke) likhoge, toh Python error nahi dega, lekin woh aapko result nahi dikhayega. Woh sirf ye batayega ki: *"Yeh ek function object hai jo yahan stored hai"*. Kyunki aapne use "kaam karne ka order" (`()`) nahi diya.
