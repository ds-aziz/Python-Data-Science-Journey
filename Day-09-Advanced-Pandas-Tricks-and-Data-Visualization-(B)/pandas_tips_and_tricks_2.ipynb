## - What every column means in kashti dataset?


Acha ji, Titanic dataset (jise Python ki `seaborn` library mein **kashti** ke naam se bhi load kiya jata hai) ko samajhna bohot aasan hai. Yeh asal mein 1912 mein doobne wale mashhoor Titanic jahaz ke musafiron (passengers) ka data hai.

Is dataset mein har ek column ka matlab aasan lafzon mein niche diya gaya hai:

---

### 1. **survived** (Bacha ya nahi?)

* **Matlab:** Kya woh musafir zinda bach gaya tha ya nahi?
* **Values:** * `0` = Nahi bach saka (Faut ho gaya)
* `1` = Zinda bach gaya



### 2. **pclass** (Passenger Class / Ticket ki category)

* **Matlab:** Musafir ne kis class ka ticket liya tha. Yeh uske ameer ya ghareeb hone ko bhi dikhata hai.
* **Values:** * `1` = First Class (Sab se mehnga/Ameer log)
* `2` = Second Class (Middle class)
* `3` = Third Class (Sab se sasta/Ghareeb log)



### 3. **sex** (Gender / Jins)

* **Matlab:** Musafir mard tha ya aurat.
* **Values:** `male` (Mard) ya `female` (Aurat).

### 4. **age** (Umar)

* **Matlab:** Musafir ki umar (years mein). Agar koi saal se chota bacha tha, to umar points mein hogi (jaise `0.42`).

### 5. **sibsp** (Behan-Bhai ya Mian-Biwi)

* **Matlab:** **Sib**lings (Behan-bhai) ya **Sp**ouse (Mian-biwi).
* **Tashreeh:** Jahaz par us musafir ke sath uske kitne behan-bhai ya shohar/biwi mojood magnetic the. Agar `0` hai, to iska matlab woh akela tha (bina partner ya behan-bhai ke).

### 6. **parch** (Waldein ya Bachay)

* **Matlab:** **Par**ents (Ami-Abu) ya **Ch**ildren (Bachay).
* **Tashreeh:** Jahaz par us musafir ke sath uske kitne bachay ya maa-baap mojood the.

### 7. **fare** (Kiraya)

* **Matlab:** Ticket ki qeemat (Pound sterling mein). Kis musafir ne safar ke liye kitne paise diye the.

### 8. **embarked** (Kahan se jahaz par sawaar hue? - Code)

* **Matlab:** Woh shehar jahan se musafir kashti (Titanic) par baitha tha.
* **Values:** * `C` = Cherbourg
* `Q` = Queenstown
* `S` = Southampton



---

### Kuch Extra Columns (Jo Python khud se add kar deta hai)

Agar aap `sns.load_dataset('titanic')` use kar rahe hain, to usmein kuch duplicate ya milti-julti columns bhi hote hain jo data ko aasan banane ke liye hote hain:

* **class:** Yeh exact `pclass` jaisa hi hai, bas yeh numbers ki jagah text mein hota hai (`First`, `Second`, `Third`).
* **who:** Musafir kon tha? (`man` = mard, `woman` = aurat, `child` = bacha).
* **adult_male:** Kya woh baaligh mard tha? (`True` ya `False`).
* **deck:** Jahaz ka kaunsa hissa ya floor tha (jaise `A`, `B`, `C`, `D` floors hote hain).
* **embark_town:** `embarked` ka mukammal naam (jaise `C` ki jagah `Cherbourg` likha hoga).
* **alive:** `survived` ka text roop (`yes` matlab zinda, `no` matlab nahi bach saka).
* **alone:** Kya musafir bilkul akela safar kar raha tha? (`True` matlab akela tha, `False` matlab sath parivar tha).

## - How many datasets are there in sns.load_datasets? and how to fetch?


`sns.get_dataset_names()` function ka istemal kar ke aap un tamam datasets ki list dekh sakte hain jo Seaborn library mein pehle se majood hain.

Chaliye aapke dono sawalat ka jawab details ke sath dekhte hain:

---

### 1. Kitne datasets hain? (How many datasets?)

Seaborn ke paas **90 se zyada (approx 95+) builtin datasets** hain. Yeh tadaad waqt ke sath thodi bohot tabdeel hoti rehti hai kyunke nayi updates mein naye datasets add kar diye jaate hain.

In mein sab se mashhoor datasets yeh hain:

* **`titanic`**: Jis par humne pehle baat ki (passengers ka data).
* **`iris`**: Phoolon (flowers) ki types ka data (Machine Learning seekhne ke liye sab se pehle yeh use hota hai).
* **`tips`**: Hotel mein khana khanay walon ki tips aur bill ka data.
* **`planets`**: Khojay gaye naye planets (exoplanets) ka data.
* **`flights`**: Saal aur mahine ke hisab se hawai jahaz ke musafiron ki tadaad.

---

### 2. Dataset ko Fetch (Load) kaise karein?

Inko fetch karne ke liye aapko pehle `seaborn` library import karni hoti hai, aur phir do aasan functions ka istemal hota hai:

#### A. Tamam datasets ke naam dekhne ke liye:

Agar aap dekhna chahte hain ke kaun kaun se naam available hain, to yeh code likhein:

```python
import seaborn as sns

# Tamam datasets ki list dekhne ke liye
all_names = sns.get_dataset_names()
print(all_names)
print(f"\nTotal datasets available: {len(all_names)}")

```

#### B. Kisi ek dataset ko load karne ke liye:

Jab aapko naam pata chal jaye, to aap `sns.load_dataset()` ke andar us ka naam short quotes `''` mein likh kar use fetch kar sakte hain:

```python
import seaborn as sns

# 1. Tips wala dataset load karein
tips_data = sns.load_dataset('tips')

# 2. Data ke shuru ke 5 rows dekhne ke liye
print(tips_data.head())

```
