# EDA (Exploratory Data Analysis):

## Exploratory Data Analysis (EDA) is the foundational data science process of investigating raw datasets to summarize their main characteristics, uncover hidden patterns, spot anomalies, and check assumptions using visual and statistical methods. It ensures data quality before building predictive machine learning models.

1. family_size column df me add karna hai

```python
# 1. Total family size calculate krna (including the passenger themselves)
df['family_size'] = df['sibsp'] + df['parch'] + 1

# 2. Yeh dekhne ke liye ke data kaisa dikh raha hai
print(df[['sibsp', 'parch', 'family_size']].head())
```

```python
df.columns
```

```python
df.head()
```

```python
# Har family size ke mutabiq survival rate check krna
family_survival = df.groupby('family_size')['survived'].mean() * 100
print("Family Size ke mutabiq Bachnay ka Percentage (%):")
print(family_survival)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Graph banana jo dikhaye ke kis family size me kitne log bache aur kitne nahi
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='family_size', hue='survived')

plt.title('Family Size vs Survival (0 = No, 1 = Yes)')
plt.xlabel('Family Size')
plt.ylabel('Passengers ka Count')
plt.legend(title='Survived', labels=['Died', 'Survived'])
plt.show()
```
