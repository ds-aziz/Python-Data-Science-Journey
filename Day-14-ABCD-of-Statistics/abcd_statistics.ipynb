## 1. Statistics Intro & Data Foundations

### Statistics Intro
* **Definition:** Data ko jama karna, analyze karna, aur us se sahi faislay (decisions) karne ki science ko Statistics kehte hain.

### Data Types
Data ki do main types hoti hain:
* **Qualitative (Categorical):** Jo numbers me nahi hota (e.g., Rang, Gender, Car Model).
* **Quantitative (Numerical):** Jo numbers me hota hai (e.g., Age, Salary, Weight).
  * *Discrete:* Jinhe gina ja sake (e.g., No. of kids = 2, 3).
  * *Continuous:* Jo points me ho sake (e.g., Height = 5.8 feet).

### Levels/Scales of Measurement
* **Nominal:** Sirf naam ya categories (e.g., Male/Female).
* **Ordinal:** Categories jinme aik tarteeb/rank ho (e.g., 1st, 2nd, 3rd Class).
* **Interval:** Barabar ka gap lekin koi True Zero nahi (e.g., Temperature in Celsius).
* **Ratio:** Barabar gap aur True Zero bhi hota hai (e.g., Age, Price).

## 2. Data Collection, Sampling & EDA

### Data Collection & Data Sampling
* **Data Collection:** Raw data ko mukhtalif zraye se jama karna.
* **Data Sampling:** Poori abadi (**Population**) me se aik chota hissa (**Sample**) nikalna taake us par research ki ja sake (e.g., Random Sampling, Stratified Sampling).

### Data Visualization
* Data ko graphs aur charts ke zariye dekhna taake patterns samajh aayein.
* **Common Charts:** Bar plots, Histograms, Scatter plots, Box plots (Seaborn aur Matplotlib ke zariye).

### EDA (Exploratory Data Analysis)
* Data ko clean karna, missing values ko handle karna, outliers dhoondna, aur modeling se pehle data ki ragon (structure) ko achi tarah samajhna.

---

## 3. Descriptive Statistics

### Central Tendency (Data ka Center)
* **Mean:** Tamam values ka average (Sum of values / Total count).
* **Median:** Data ko tarteeb dene ke baad bilkul beech wali value.
* **Mode:** Data me sab se zyada dafa aane wali value.

### Variability (Data kitna phela hua hai)
* **Range:** Sab se bari value aur sab se choti value ka farq (Max - Min).
* **Variance:** Data points ka apnay Mean se average faasla (squared distance).
* **Standard Deviation:** Variance ka square root. Yeh batata hai ke aam taur par data mean se kitna door hai.

### Data Distribution
* **Normal Distribution:** Bell-shaped curve jahan Mean, Median, aur Mode barabar hotey hain.
* **Skewness:** Data ka aik taraf jhukao (Left skewed ya Right skewed).
* **Kurtosis:** Data ke curve ki peak kitni oonchi ya flat hai (Tails ki hardness).

## 4. Probability Theory & Inferential Statistics

### Probability Theory (Bayesian Theory)
* **Probability:** Kisi event ke hone ke chances.
* **Bayesian Probability (Besim Theory):** Purani maloomat (Prior knowledge) ki buniyaad par nayi prediction ko update karna.

### Inferential Statistics
* **Population vs Sample:** Poori dunya ka data vs usme se liya gaya aik representative tukda.
* **Sampling Distribution:** Bohot saare samples ke means ka distribution.
* **Confidence Interval:** Aik range (e.g., 95%) jis me hume yaqeen hota hai ke population ka asal jwab is ke darmiyan hoga.
* **Hypothesis Testing:** Aik claim ko data ke zariye check karna ke wo sach hai ya ittefaq (e.g., $H_0$: Null Hypothesis, $H_1$: Alternative Hypothesis).

### Types of Error
* **Type 1 Error ($\alpha$):** False Positive (Sachi baat ko ghalt keh dena — e.g., Aik shareef aadmi ko chor keh dena).
* **Type 2 Error ($\beta$):** False Negative (Ghalt baat ko sahi keh dena — e.g., Aik chor ko chor na manna).

### Common Statistical Tests (Comparison)
* **Univariate:** Sirf aik variable ka analysis (e.g., Age).
* **Bivariate:** Do variables ke darmiyan taluq (e.g., Age vs Salary).
* **Multivariate:** Do se zyada variables ka aik sath analysis.
* **Tests:** T-test, ANOVA, Chi-Square (Data ki type ke mutabiq check karne ke liye).

### P-values and Significance
* **P-value:** Yeh probability batati hai ke aap ka result sirf aik ittefaq (chance) hai ya haqeeqat.
* Agar **P-value < 0.05** ho, to result statistically significant hota hai (yaani ittefaq nahi hai, koi solid waja hai).
