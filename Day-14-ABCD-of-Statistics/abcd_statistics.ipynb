## Scale/levels of measurement in statistics
1. Nominal scale (Male, Female),(Dog, Cat)  nominal means jis me names ke base pr chalta hai labels or qualities.

  **No Ranking or No Ordering**

- Stat jo lagayegy:

    - Mode, frequency, ya chi square test.

1. Ordinal scale (same nominal wali khasyt hoti hai isme) but is me ek or cheez add hojati hai wo hai ranking or ordering things ya data ko ranking dena ya or order wise organize krna.
"Is me jo data hoga same nominal ka jesa he hoga but is me rank wise hoga arrange" like categorical data hoga hamare pas
