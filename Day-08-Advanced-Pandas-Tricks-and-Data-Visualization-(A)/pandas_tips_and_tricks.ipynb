# 16- Splitting a string into multiple columns

```python
# importing libraries
import pandas as pd
df = pd.DataFrame({'Name':['Aziz Baloch', 'Ammar Baba', 'Ali Hassan', 'Sara Khan'],
                   'Location':['Karachi, Pakistan', 'Lahore, Pakistan', 'Islamabad, Pakistan', 'Quetta, Pakistan']})

df
```

```python
# spilting a two columns
df.Name.str.split(' ', expand=True)
```

```python
#adding those splits into new columnCs
df[['First Name', 'Last Name']] = df.Name.str.split(' ', expand=True)
df
```

```python
# splitting a two columns
df.Location.str.split(' ', expand=True)
```

```python
#adding those splits into new columnCs
df[['City', 'Country']] = df.Location.str.split(', ', expand=True)
df
```

```python
#refine data manipulation

df = df[['First Name', 'Last Name', 'City', 'Country']]
df
```
