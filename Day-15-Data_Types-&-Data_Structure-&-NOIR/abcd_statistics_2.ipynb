## 1. Categorical Data Types (Giroh ki Shaklein)

Categorical data ka matlab hai woh data jo numbers me nahi, balki **groups ya labels** me hota hai.

*   **Nominal Data (Sirf Naam):** 
    *   **Concept:** Isme sirf naam hote hain. Koi order, ranking ya mukabla nahi hota. 
    *   **Layman Analogy:** Jaise logon ki aankhon ka rang (Black, Blue, Brown). Koi rang doosre se bada ya chota nahi hota.
    *   **Data Science Connection:** Titanic dataset me `Sex` (Male, Female) ya `Embarked` (C, Q, S) nominal data hain.
*   **Ordinal Data (Tarteeb-e-Zail):** 
    *   **Concept:** Naam ke sath-sath aik **khas tarteeb (order/rank)** bhi hoti hai.
    *   **Layman Analogy:** Jaise hotel ka feedback (Bad, Average, Good, Excellent) ya T-shirt sizes (S, M, L, XL).
    *   **Data Science Connection:** Passenger Class (`Pclass`: 1st, 2nd, 3rd) aik ordinal data hai kyunki 1st class superior hai.
*   **Boolean Data (Haan ya Naa):** 
    *   **Concept:** Binary data jisme sirf **do hi raste** hote hain.
    *   **Layman Analogy:** Switch ON (1) hai ya OFF (0)? Aap paas hain (True) ya fail (False)?
    *   **Data Science Connection:** Titanic me `Survived` (0 = No, 1 = Yes) bilkul pure boolean data hai.

## 2. Number of Variables (Kitni Cheezon Par Nazar Hai?)

Hum aik waqt me kitne badalte hue nikaat (variables) ka muazna kar rahe hain, isko statistics me yun baanta jata hai:

*   **Univariate Data:** 
    *   **Meaning:** `Uni` matlab Aik. Jab hum sirf **aik hi variable** ka test ya analysis karte hain.
    *   **Example:** Iris dataset me sirf phoolon ki patti ki lambai (`PetalLength`) ka graph banana.
*   **Bivariate Data:** 
    *   **Meaning:** `Bi` matlab Do. Jab **do variables** ko aapas me jor kar unka rishta dekha jaye.
    *   **Example:** Garmi ka temperature badhne se Ice Cream ki sale ka badhna. (Temperature vs Sales).
*   **Multivariate Data:** 
    *   **Meaning:** `Multi` matlab Bohot Saare. Jab **do se zayada variables** ko aik sath study kiya jaye.
    *   **Example:** Kisi ghar ki keemat nikalna jisme uski location, kamray, square feet aur naya/purana hona sab aik sath dekha jaye.

## 3. Data Structure (Tameez Shuda vs Bikhra Hua)

Computer ko data feed karne ke liye uske dhanchay (structure) ko samajhna zaroori hai.

*   **Structured Data:** 
    *   Yeh bilkul shareef aur tameez shuda data hai jo **Rows and Columns (Tables)** me fixed hota hai. Jaise Excel sheets, SQL databases, ya Pandas DataFrame.
*   **Unstructured Data:** 
    *   Yeh bikhra hua data hai jiska koi fixed table nahi hota. Computer isko seedha nahi samajh sakta, isko pehle process karna parta hai. Isme shamil hain:
        *   `Textual Data / Text files`: Paragraphs, book chapters, ya tweets.
        *   `Multimedia`: Audio, Video, aur Images.
        *   `Webpages`: HTML pages jahan har cheez bikhri hoti hai.
*   **Semi-Structured Data:** 
    *   Jo poori tarah bikhra hua bhi nahi hota aur poori tarah table bhi nahi hota. Isme tags hote hain. Jaise **JSON** aur **XML** files.

## 4. Measurement Scales (The NOIR Framework)

Yeh pooray statistics ka sab se ahem scale hai. Isko yaad rakhne ki short trick hai **NOIR**:

| Scale | Features | Layman Example | Data Science Example |
| :--- | :--- | :--- | :--- |
| **N**ominal | Sirf Category (No Math, No Order) | Marital Status (Single/Married) | Country names |
| **O**rdinal | Category + Order (No Fixed Distance) | Race Positions (1st, 2nd, 3rd) | Customer Satisfaction |
| **I**nterval | Order + Equal Distance (No True Zero) | Temperature ($0^\circ\text{C}$ does not mean no heat) | IQ Score |
| **R**atio | Order + Distance + **True Zero (0 = Nothing)** | Weight (0 kg means no weight), Salary | Ticket `Fare`, `Age` |

### 💡 Crucial Concept: Interval vs Ratio
*   **Interval me 0 aik nishan hota hai, khatam hona nahi.** Agar temperature $0^\circ\text{C}$ hai, to thand hoti hai, garmi gayab nahi hoti.
*   **Ratio me 0 ka matlab hai absolute khatam!** Agar aapki jeb me $0$ rupees hain, to iska matlab sach me koi paisa nahi hai.
