## Step 1: Pehle check karo missing values hain kitni?

Sabse pehle dataset load karke missing values ka status check karte hain.

```python
import pandas as pd
import numpy as np

# Titanic dataset load karein (agar aapke paas local file hai)
df = pd.read_csv('titanic.csv')

# Missing values ka percentage dekhne ke liye
print((df.isnull().sum() / len(df)) * 100)

```

**Output mein aapko generally yeh scene dikhega:**

* `Age`: ~20% missing values.
* `Cabin`: ~77% missing values (Bohot zyada!).
* `Embarked`: ~0.2% missing values (Sirf 2 rows missing).

---

## Step 2: Har datatype aur column ko handle karna

### 1. Numeric Column: `Age` (~20% Missing)

`Age` ek numeric column hai. Kyunki missing data sirf 20% hai, toh hum rows delete nahi karenge. Hum isse **Median** se fill karenge kyunki age mein kuch bache aur kuch bohot boorhe log bhi the (outliers).

* **Aasan Approach (Overall Median):** Poore dataset ka jo median age hai, usse fill kar do.

```python
age_median = df['Age'].median()
df['Age'].fillna(age_median, inplace=True)

```

* **Smart Approach (Advanced EDA):** Sirf blind median lagane ke bajaye, logon ke Naam ke sath lage Title (`Mr.`, `Mrs.`, `Miss`, `Master`) ke hisab se group karke median nikalo. Kyunki `Master` (bache) ki median age kam hogi aur `Mr.` ki zyada.

```python
# Naam se Title nikalne ka shortcut
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Har Title ke hisab se Age ka median fill karna
df['Age'] = df.groupby('Title')['Age'].transform(lambda x: x.fillna(x.median()))

```

---

### 2. Categorical Column (High Missing): `Cabin` (~77% Missing)

`Cabin` ek object/categorical column hai. Ismein **77% data missing** hai. Agar hum isko mode se fill karenge toh data corrupt ho jayega.

* **Best Approach:** Is missing status ko hi ek nayi category bana do (`Unknown` ya `M`), kyunki ship par jinka cabin number nahi likha tha, high chances hain ki wo lower class (Third Class) ke musafir the. Yani missing hona bhi ek statistical information hai!

```python
df['Cabin'].fillna('Unknown', inplace=True)

```

---

### 3. Categorical Column (Low Missing): `Embarked` (0.2% Missing)

`Embarked` (jahan se musafir ship par sawaar hue) mein sirf **2 rows missing** hain.

* **Best Approach:** Kyunki yeh categorical data hai aur missing values bohot hi kam hain, toh hum **Mode** (jo city sabse zyada baar repeat hui hai) use karenge. Titanic par sabse zyada log 'S' (Southampton) se charhe the.

```python
# Sabse zyada repeat hone wali city (Mode)
mode_embarked = df['Embarked'].mode()[0]

# Fill kar do
df['Embarked'].fillna(mode_embarked, inplace=True)

```
