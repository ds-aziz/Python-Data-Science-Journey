## Bias in data:
In Data Science and Statistics, **Bias in Data** (ya Data Bias) tab hota hai jab aapka collected data real-world population ko sahi se represent nahi karta. Agar data hi biased hoga, to us par banaye gaye graphs, insights, aur Machine Learning models bhi hamesha ghalat ya ek-tarfa (skewed) results dein ge.

GIGO principle yahan bilkul fit baithta hai: **Garbage In, Garbage Out**.

Data bias ki kuch ahem aur aam qasmein (types) ye hain:
### 1. Sampling Bias
occurs when a not representative of a population. iska matlb ye hai ke jo sample data hamne uthaya wo population ko represent he nh kar rha hai.

* **Example:** Mere pas hai Section C class jisme se mujhe CR banana hai hai class ka, but me Section D me se CR bana deta ho iska mtlb ke mene Section c ko represent he nh krwa rha or na he jo CR mene banaya wo Section C se nh hai balky Section D ko represent kar rha hai.

### 2. Selection Bias

Ye tab hota hai jab data collect karte waqt sample population ko random tareeqe se select nahi kiya jata. Kuch logon ya cheezon ke select hone ke chances doosron se zyada hote hain.

* **Example:** Agar aap pure Pakistan ki internet usage par research karna chahein, lekin survey ka data sirf Karachi aur Lahore ke posh areas se collect karein, to ye selection bias hoga kyunki gaon (rural areas) ka data isme shamil hi nahi.

### 3. Survivorship Bias

Isme aap sirf un cheezon ya logon ka data analyze karte hain jo kisi mushkil ya process se guzar kar **bach gaye** (survive kar gaye), aur jo fail ho gaye unka data aap ignore kar dete hain.

* **Example:** Sirf un startups ka data study karna jo aaj billion-dollar companies hain, aur un 90% startups ka data chor dena jo pehle saal hi band ho gaye. Isse aapko lagay ga ke startup kholna bohot aasan hai.

### 4. Reporting / Response Bias

Ye tab hota hai jab log hamesha sach ya mukammal baat nahi batate, balki wahi jawab dete hain jo unhe lagta hai ke sunne wala sunna chahta hai, ya phir sirf wahi log jawab dete hain jo bohot zyada motivated hote hain.

* **Example:** Kisi product ke online reviews dekhna. Kashti ya kisi product ke reviews zyadatar wahi log likhte hain jo ya to zaroorat se zyada khush hote hain ya zaroorat se zyada gusse me hote hain. Darmiyan wale log aksar review nahi likhte.

### 5. Historical Bias

Jo insani ta'assub, ghalat fehmiyan, ya stereotypes pehle se samaj me chali aa rahi hain, jab woh purana data me record ho jati hain aur hum wahi data model ko de dete hain.

* **Example:** Agar past me kisi field (jaise Tech) me sirf mardoon ko zyada hire kiya gaya ho, to naya AI model purana data dekh kar ye samajh lega ke mard hi is job ke liye behtar hain aur women applicants ko filter out karne lagega.

---

### Data Bias Se Bachne Ka Tariqa?

1. Improve sampling methods (randomize sampling)
2. refined research designs (like blinding techniques or control groups)
3. data collection (vigillance or standadize)
4. analytical awareness (use statistics testing for check, peer review)
5. ongoing education and awareness.(courses kary, books parh kr jany biasness ke bary me, or critical thinking ko improve krna chahye)

# Population vs Sample in Data Science:

Data science aur statistics mein **Population** aur **Sample** ka farq samajhna sab se pehla aur sab se zaroori step hai. Is ko asan lafzon mein samajhte hain:

---

## 1. Population (Puri Ki Puri Dunya)
Jis poore group ke baare mein aap koi information ya data jama karna chahte hain, usay **Population** kehte hain. Yeh koi zaroori nahi ke sirf insaan hon; yeh total data points, total cars, ya kisi e-commerce site ke saare customers bhi ho sakte hain.

> **Example:** Agar aap dekhna chahte hain ke Titanic (`kashti`) ke saare musafiron ki average age kya thi, to Titanic par majood **saare ke saare musafir (all passengers)** aap ki population hain.

## 2. Sample (Us Ka Aik Chota Tukra)
Population mein se select kiya gaya aik chota hissa (subset) jis par aap actual mein research ya coding karte hain, usay **Sample** kehte hain. 

> **Example:** Saare musafiron ka data nikalne ki bajaye agar aap sirf **100 random musafiron** ka data select kar ke un par analysis karte hain, to woh 100 log aap ka sample hain.

---

## Dono Mein Main Farq (Comparison Table)

| Feature | Population | Sample |
| :--- | :--- | :--- |
| **Definition** | Poora ka poora group jise aap study karna chahte hain. | Population ka aik chota hissa jo analyze kiya jata hai. |
| **Size** | Bohat bara hota hai (aksar is ka data nikalna mushkil ya expensive hota hai). | Chota aur manageable hota hai. |
| **Python Context** | Poora raw dataset ya real-world ka total data. | `.sample(n=100)` ka use kar ke jo chota dataframe hum nikalte hain. |
| **Statistical Term**| Is ki characteristics (jaise Mean) ko **Parameter** kehte hain. | Is ki characteristics ko **Statistic** kehte hain. |

---

## Ek Real-Life Example 🍲

Farz karein aap ke ghar mein **Biryani** pakk rahi hai:

* **Population:** Degh mein majood **saari ki saari biryani** population hai.
* **Sample:** Aap check karne ke liye chamach mein jo **aik nawaala (bite)** lete hain ke namak aur mirch theek hai ya nahi, woh sample hai.

> *Aap poori degh nahi khaate check karne ke liye, sirf aik chamach se poori degh ka andaza laga lete hain!* Data science mein bhi hum sample par analysis kar ke poori population ka andaza lagate hain.

## Pandas (Python) mein isay kaise dekhte hain?

Jab aap kisi dataset par kaam kar rahe hon (jaise `kashti` ya Titanic data), to aap poore dataset ko population maan sakte hain, aur agar aap ko us mein se random sample nikalna ho to Pandas mein aik simple function hota hai:

```python
import seaborn as sns

# Poora dataset load kiya (Yeh hamari Population ki tarah hai)
df = sns.load_dataset('titanic')

# Is mein se 50 random rows nikal lein (Yeh hamara Sample ban gaya)
sample_df = df.sample(n=50)

print(sample_df.shape)  # Output: (50, 15)
